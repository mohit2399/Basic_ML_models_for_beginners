```python
import gym
env = gym.make('CartPole-v0')
print(gym.__version__)

env.reset()
for _ in range(500):
    env.render()
    obs, rew, done, info = env.step(env.action_space.sample()) # take a random action
    if done:
       env.reset()
env.close()
```
